Navigating to the Zip file

```python
zip_path = "/content/drive/MyDrive/Potholes/Kuwait Potholes.v1i.yolov8.zip"  # Provide path of the zip file
```

Unzipping the file

```python
import zipfile

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/content/potholes_dataset')       ##a new file named 'potholes_dataset' will be created
```

Training the model

```python
from ultralytics import YOLO

# Load YOLOv8 nano model (smallest and fastest)
model = YOLO('yolov8n.pt')

# Train the model
model.train(data='/content/potholes_dataset/data.yaml', epochs=20, imgsz=640)
```

Predictions

```python
results = model.predict(source='/content/potholes_dataset/valid/images', save=True)
```

View Results
